```python
import re
import os
import sys
import csv
import time
import random
import string
import requests
import datetime
```

```python
def build(terms_id, dir_path):
    path = "{}/danmu3.txt".format(str(dir_path))
    with open(path, 'w', newline='') as f:
        for key in terms_id:
            try:
                f.write(str(terms_id[key]))
                f.write('\n')
            except UnicodeEncodeError:
                break
```

```python
def danmu_crawler(danmu_url, headers, dir_path):
    if not os.path.exists(dir_path):
        os.makedirs(dir_path)
    response = requests.get(danmu_url, headers=headers)
    array1 = dict()
    if response.status_code == 200:
        danmu_xml = response.content.decode("utf8","ignore")
        res_name = com.findall(danmu_xml)
        for item in res_name:
            if item[0] not in array1:
                array1[item[0]] = item[1]
        if len(res_name) < 2:
            print('查不到了呢')
            print(danmu_xml)
    build(array1, dir_path)
```

```python
com = re.compile(r",(\d+)\">(\w+)</d>", re.M)
com_info = re.compile(r"\"cid\":(\d+),.*?,\"long_title\":\"([^\"]+)\"", re.M)
com_season = re.compile(r"'(\d+)':'(\w+)'", re.M)

## "season_id":26801,"season_type":1,"stat":{"danmaku":7126473,"follow":7844823,"view":443862480},"title":"鬼灭之刃",
```

```python
seasons = dict()
reader = csv.reader(open('season_id.csv', "r"), delimiter=':')
for item in reader:
    #print(item)
    seasons[item[0].replace("\'", "").strip()] = item[1].replace("\''", "").replace(",", "").strip()
```

```python
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36'}
fanju_url = 'https://api.bilibili.com/pgc/web/season/section?season_id='
```

```python
f=open(r'cookies.txt','r')
cookies={}
for line in f.read().split(';'):
    item = line.strip().split('=')
    cookies[item[0]]=item[1]
```

```python
for s in seasons:
    response_info = requests.get(fanju_url+str(s), headers=headers,cookies = cookies)
    info_xml = response_info.content.decode("utf8","ignore")
    info_name = com_info.findall(info_xml)
    fanju_name = seasons[s]
    fanju_name = str(fanju_name).replace(' ', '').replace('//','-')
    fanju_name = re.sub("[\s+\.\!\/_,$%^*(+\"\']+|[+！<>，/。？?!,.:：||、~@#￥%……&*（）]+", "",fanju_name)
    cnt = 0
    print(fanju_name)
    for episode in info_name:
        cnt += 1
        danmu_url = 'http://comment.bilibili.com/{}.xml'.format(str(episode[0]))
        ep = str(episode[1]).replace(' ', '').replace('//','-')
        ep = re.sub("[\s+\.\!\/_,$%^*(+\"\']+|[+！，<>//。？||?!,.:：、~@#￥%……&*（）]+", "",ep)
        if len(info_name)> 999:
            dir_path = "动漫/{}/{}".format(fanju_name, "{:0>4d}".format(cnt)+'_'+ep)
        elif len(info_name)> 99:
            dir_path = "动漫/{}/{}".format(fanju_name, "{:0>3d}".format(cnt)+'_'+ep)
        else:
            dir_path = "动漫/{}/{}".format(fanju_name, "{:0>2d}".format(cnt)+'_'+ep)
        danmu_crawler(danmu_url, headers, dir_path)
```
