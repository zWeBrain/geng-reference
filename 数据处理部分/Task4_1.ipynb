```python
def shorten(s):
    ss= ''.join(reversed(s))
    k = -1
    j = 0
    array = [-1]
    plen = len(ss)
    zeros = 0
    nonze = 0
    while j < plen:
        if k == -1 or ss[j] == ss[k]:
            k += 1
            j += 1
            if k == 0:
                if j != zeros + 1:
                    loop = (int)(nonze/zeros) -1
                    return cut_string(s, plen) if loop<1 else cut_string(s, plen - zeros*loop)
                zeros += 1
            else:
                nonze += 1
            array.append(k)
        else:
            k = array[k]
    return cut_string(s, plen) if zeros==plen else (cut_string(s, zeros) if zeros>1 else cut_string(s, zeros+1)) 
def cut_string(s, lens):
    return s[0: lens]
```

```python
print(shorten("23333333"))
print(shorten("wryyyyyyyy"))
print(shorten("ohhhhhhhhhhh"))
print(shorten("义勇啊啊啊啊啊啊"))
print(shorten("木大木大木大木大木大"))
print(shorten("你们有毒吧哈哈哈哈哈哈"))
print(shorten("得得得得得得得得得得得得得得得得得"))
```
